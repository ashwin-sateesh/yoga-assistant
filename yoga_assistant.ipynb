### Preprocessing


```python
# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_path):
    document = fitz.open(pdf_path)
    text = ""
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        text += page.get_text()
    return text

# Function to generate prompt and completion using GPT-3.5
def generate_prompt_completion(chunk):
  prompt = ChatPromptTemplate.from_template(
           "Create appropriate a question or a prompt from the following text: \
            \n\n \
            {chunk} \n\n \
            Question:"
            )
  llm_mod = "gpt-3.5-turbo"
  llmx = ChatOpenAI(temperature=0.9, model=llm_mod)
  chain = LLMChain(llm=llmx, prompt=prompt, verbose=False)
  question = chain.run(chunk)

  return {"prompt": question, "completion": chunk}


# Function to format data for fine-tuning
def format_for_fine_tuning_with_gpt(texts, chunk_size=500):
    data = []
    for text in texts:
        # Split text into chunks of a specific size
        chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
        for chunk in chunks:
            prompt_completion_pair = generate_prompt_completion(chunk)
            data.append(prompt_completion_pair)
            time.sleep(1)  # Add a delay to avoid rate limiting
    return data

# Function to format data for fine-tuning
def format_for_fine_tuning_with_gpt1(text, num_chunks=10):
    data = []

    # Calculate chunk size to ensure text is divided into exactly num_chunks
    chunk_size = max(1, len(text) // num_chunks)

    # Split text into chunks of calculated size
    chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]

    # Ensure we have exactly num_chunks by merging last chunk if needed
    if len(chunks) > num_chunks:
        chunks = chunks[:num_chunks-1] + [''.join(chunks[num_chunks-1:])]

    for chunk in chunks:
        prompt_completion_pair = generate_prompt_completion(chunk)
        data.append(prompt_completion_pair)
        time.sleep(1)  # Add a delay to avoid rate limiting

    return data
```

### Queue based Finetuning gpt-3.5 with web scraping and update


```python
# Function to extract URL and combined text from the prompt
def extract_url_and_text(prompt):
  # Match the first occurrence of a URL in the prompt
  url_match = re.search("(?P<url>https?://[^\s]+)", prompt)
  if url_match:
    url = url_match.group("url")
    # Extract text before and after the URL
    before_url = prompt[:url_match.start()].strip()
    after_url = prompt[url_match.end():].strip()
    combined_text = before_url + " " + after_url  # Combine text before and after URL
    return url, combined_text
  else:
    return url_match, prompt

def scrape_content(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    # Extract relevant content from the webpage using BeautifulSoup
    content = soup.get_text()  # Simplified example; adjust as per your needs
    return content

client = OpenAI()

def create_tuning_data(prompts, completions, file_name, dir):
    tuning_data = []

    # Iterate over prompts and completions to create message dictionaries
    for prompt, completion in zip(prompts, completions):
        system_message = {
            "role": "system",
            "content": "You are a knowledgeable yoga assistant. Answer questions based on the provided yoga related text."
        }
        user_message = {
            "role": "user",
            "content": prompt
        }
        assistant_message = {
            "role": "assistant",
            "content": completion
        }

        # Append the messages to the data list
        tuning_data.append({"messages": [system_message, user_message, assistant_message]})

    # Write the data to a JSONL file
    jsonl_path = os.path.join(dir, file_name)
    with open(jsonl_path, "w") as jsonl_file:
        for item in tuning_data:
            jsonl_file.write(json.dumps(item) + '\n')

    return jsonl_path

def get_status(job_id):
    # Retrieve events for the fine-tuning job
    events = client.fine_tuning.jobs.list_events(
                fine_tuning_job_id=job_id,
                limit=2
                )
    # Iterate over events to find the model name and status
    status = None
    for event in events.data:
      if 'The job has successfully completed' in event.message:
            status = 'completed'
      else:
            status = 'not completed'
      return status


def get_model_name(job_id, previous_model_name = 'ft:gpt-3.5-turbo-0125:personal::9ejYHbZX'):
  # Retrieve events for the fine-tuning job

  events = client.fine_tuning.jobs.list_events(
              fine_tuning_job_id=job_id,
              limit=2
              )
  #events = openai.FineTuning.list_events(fine_tuning_job_id=job_id, limit=2)
  # Iterate over events to find the model name and status
  model_name = previous_model_name
  for event in events.data:
    if 'New fine-tuned model created' in event.message:
        # Extract the new model name from the message
        new_model_name = event.message.split(': ')[1]
        model_name = new_model_name
        return new_model_name
    else:
      return model_name

from queue import Queue
model_tuning_queue = Queue()
```

```python
def update_model_with_content(content = None, previous_job_id = 'ftjob-rfbbi3Zb5PiGpqQ4eSx3Aohq',
                              previous_model_name = 'ft:gpt-3.5-turbo-0125:personal::9ejYHbZX',
                              previous_file_id = 'file-QTX0Btg4wu6bS9Iv3nT4SrOF',
                              previous_llm_model = 'gpt-3.5-turbo' ,
                              model_tuning_queue = model_tuning_queue,
                              url= None):

  if url != None:
    data = format_for_fine_tuning_with_gpt1(content)
    prompts, completions = [],[]
    for entry in data:
      prompts.append(entry['prompt'])
      completions.append(entry['completion'])

    jsonl_path = create_tuning_data(prompts, completions, "yoga_bot_data_update.jsonl", "/content/drive/MyDrive/yoga_bot")

    #client = OpenAI()
    file_details = client.files.create(
                file=open(jsonl_path, "rb"),
                purpose="fine-tune"
                )
    model_tuning_queue.put(file_details.id)
    #recent_file_id = model_tuning_queue.get()
    if get_status(previous_job_id) == 'completed':
      file_id = model_tuning_queue.get()
      previous_file_id = file_id

      response = client.fine_tuning.jobs.create(
            training_file= file_id,
            model= previous_model_name,
            hyperparameters={
            "n_epochs":2,
            "learning_rate_multiplier":0.1
            }
    )

    else:
      file_id = previous_file_id

    job_id = response.id
    status = get_status(job_id)
    if status == 'completed':
      job_id = response.id
      previous_job_id = job_id
      previous_model_name = get_model_name(job_id, previous_model_name)
    else:
      job_id = previous_job_id
      previous_model_name = get_model_name(job_id, previous_model_name)

    current_fine_tuned = client.fine_tuning.jobs.retrieve(job_id)
    llm_model = current_fine_tuned.fine_tuned_model
    previous_llm_model = llm_model
    llm = ChatOpenAI(temperature=0.9, model=llm_model)


  else:
    file_id = previous_file_id
    job_id = previous_job_id
    previous_model_name = get_model_name(job_id, previous_model_name)
    llm_model = previous_llm_model
    llm = ChatOpenAI(temperature=0, model=llm_model)

  latest_model_details = [file_id, job_id, previous_model_name]
  return llm, latest_model_details
```

## Inference

### GPT Inference

```python
def generate_response_from_link(inp_query, scraped_content, llm_model):
    prompt1 = ChatPromptTemplate.from_template(
        """Using the content below, give appropriate response to the following query:
           Content:
           {content}

           Query:
           {query}

           Response:"""
    )
    chain = LLMChain(llm=llm_model, prompt=prompt1, verbose=False)
    new_response = chain.run(content = scraped_content, query = inp_query)
    return new_response


# Function to generate response after scraping using GPT-3.5
def generate_response_normal(inp_query, llm_model):
    prompt2 = ChatPromptTemplate.from_template(
        """Give appropriate yoga related response to the following query:

            Query:
           {query}

           Response:"""
    )
    chain = LLMChain(llm=llm_model, prompt=prompt2, verbose=False)
    new_response = chain.run(inp_query)
    return new_response
```

### Query Classification

```python
visualization_keywords = [
    "show", "display", "visualize", "image", "picture",
    "illustrate", "depict", "render", "sketch", "draw", "demonstrate",
    "exhibit", "present", "graph", "diagram", "chart", "map",
    "photograph", "snapshot", "view", "screen", "projection",
    "representation", "portrait", "photo", "painting", "represent",
    "rendering", "visual", "presentation", "figure", "frame",
    "scene", "screenshot", "capture", "layout", "design"
]

def classify_query(query):
    for keyword in visualization_keywords:
        if re.search(rf"\b{keyword}\b", query, re.IGNORECASE):
            return "image"
    return "text"
```

### Full Model Inference

```python
def respond(input_query):
  url, text = extract_url_and_text(input_query)
  query_asked = classify_query(input_query)

  if url == None and query_asked == "text":
    llm, latest_model_details = update_model_with_content(url= url)
    new_response = generate_response_normal(input_query, llm)

  elif query_asked == "image":
     new_response = StableDiffusionInference(input_query)

  else:
    scraped_content = scrape_content(url)
    llm, latest_model_details = update_model_with_content(content=scraped_content, url= url)
    new_response = generate_response_from_link(input_query, scraped_content, llm)
  return new_response
```

### Evaluation

```python
def calculate_perplexity(response, model, tokenizer):
    encodings = tokenizer(response, return_tensors='pt')
    input_ids = encodings.input_ids
    with torch.no_grad():
        outputs = model(input_ids, labels=input_ids)
        loss = outputs.loss
        perplexity = torch.exp(loss).item()
    return perplexity
```

```python
import nltk
from nltk.chat.util import Chat, reflections

# Pairs is a list of patterns and responses.
pairs = [
    [
        r"hi|hello|hey",
        ["Hello, I'm your Yoga Assistant. How can I assist you today?",]
    ],
    [
        r"quit",
        ["Goodbye! If you have more questions in the future, don't hesitate to ask.",]
    ],
]

# This is a simple reflection function that can be used to flip a few pronouns
# e.g. 'I am' becomes 'you are'
reflections = {
    "i am": "you are",
    "i was": "you were",
    "i": "you",
    "i'm": "you are",
    "i'd": "you would",
    "i've": "you have",
    "i'll": "you will",
    "my": "your",
    "you are": "I am",
    "you were": "I was",
    "you've": "I have",
    "you'll": "I will",
    "your": "my",
    "yours": "mine",
    "you": "me",
    "me": "you"
}

# Create your own Chatbot
yoga_assistant = Chat(pairs, reflections)

def yoga_assistant_chat():
    print("Yoga Assistant\n---------")
    print("Hello! I'm your Yoga Assistant. I can help you with yoga-related questions.\nType 'quit' to leave the chat.")

    while True:
        user_input = input("You: ")
        if user_input.lower() == 'quit':
            print("Yoga Assistant: Goodbye!")
            break

        response = respond(user_input)
        if response.endswith(".png"):
            # If the response is an image path, display the image
            image = Image.open(response)
            display(image)  # Use IPython.display to show the image
            print("Yoga Assistant: Here's the image related to your query.")
        else:
            # Otherwise, format and display the text response
            display(Markdown(response))
```
